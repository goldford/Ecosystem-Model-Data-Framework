### exploring length at age and growth in fresh vs marine for Pac salmon

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
df = pd.read_csv('D:\\temp_ewe\\SizeAtAgesPacSalmon_v1.csv')

# Convert 'month_after_fry_emergence' column to numeric
df['month_after_fry_emergence'] = pd.to_numeric(df['month_after_fry_emergence'], errors='coerce')

# Filter out rows with 0 values in relevant columns
df1 = df.copy()
df1 = df1[(df1['FL_avg'] != 0) | (df1['FL_min'] != 0) | (df1['FL_max'] != 0) | (df1['FL_SD'] != 0)]

# Convert 0 values to NaN
df1.replace(0, pd.NA, inplace=True)

# Create a figure for each unique 'species_and_subtype'
species_subtypes = df1['species_and_subtype'].unique()


for subtype in species_subtypes:
    # Filter the data for the current subtype
    subset = df1[df1['species_and_subtype'] == subtype]
    
    # Create a subplot
    fig, ax = plt.subplots()

    # Iterate over each record in the subset
    for index, row in subset.iterrows():
        # Check if only 'FL_avg' is available
        if pd.notna(row['FL_avg']) and pd.isna(row['FL_min']) and pd.isna(row['FL_max']) and pd.isna(row['FL_SD']):
            # Scatter plot with only average value
            ax.scatter(row['month_after_fry_emergence'], row['FL_avg'], marker='o', label='Average')
        elif pd.notna(row['FL_SD']):
            # Manually compute quartiles
            q1 = row['FL_avg']-0.675*row['FL_SD']
            q3 = row['FL_avg']+0.675*row['FL_SD']

            # Manually define boxplot values
            box_values = {
                'med': row['FL_avg'],
                'q1': q1,
                'q3': q3,
                'whislo': row['FL_min'],
                'whishi': row['FL_max'],
            }

            # Box plot with manually defined values
            ax.bxp([box_values], positions=[row['month_after_fry_emergence']], widths=0.6, showfliers=False)
        else:
            # Whisker plot with manually defined values
#             ax.plot(row['month_after_fry_emergence'], row['FL_avg'], marker='o', label='Average')
#             ax.plot(row['month_after_fry_emergence'], row['FL_min'], marker='o', label='Min')
#             ax.plot(row['month_after_fry_emergence'], row['FL_max'], marker='o', label='Max')
            
            # Manually define boxplot values
            box_values = {
                'med': row['FL_avg'],
                'q1': row['FL_avg'],
                'q3': row['FL_avg'],
                'whislo': row['FL_min'],
                'whishi': row['FL_max'],
            }
            
             # Box plot with manually defined values
            ax.bxp([box_values], positions=[row['month_after_fry_emergence']], widths=0.6, showfliers=False)
       
            
            
    # for label of fresh / mar
    print(subtype)
    subset2 = df[df['species_and_subtype'] == subtype]
    loop_n = 0
    for index, row in subset2.iterrows():
        # Check for habitat change
        if loop_n > 0:
            if index > 1 and subset2.at[index, 'habitat'] != subset2.at[index - 1, 'habitat']:
                # Add vertical dashed line at habitat change
                x_mo_marentry = row['month_after_fry_emergence']
                FL_avg_marentry = row['FL_avg']
                FL_SD_marentry = row['FL_SD']
                
                ax.axvline(x=x_mo_marentry, linestyle='--', color='gray', label=None)

                # Add labels on the sides of the line
                ax.text(row['month_after_fry_emergence'], ax.get_ylim()[1]*0.7, 'fresh', ha='right', va='bottom', rotation=90)
                ax.text(row['month_after_fry_emergence'], ax.get_ylim()[1]*0.7, 'marine', ha='left', va='bottom', rotation=90)
        loop_n += 1
        
    ax.set_xlabel('Month (after emergence)')
    ax.set_ylabel('Length')
    #ax.legend()

    ax.set_title(f'{subtype} - Plot')
    
    # Try to add trend line + cred int
    # will use 12 months after marine entry +/- 2months
    months_later = x_mo_marentry + 11
    if (
        months_later in subset2['month_after_fry_emergence'].values and
        subset2.loc[subset2['month_after_fry_emergence'] == months_later, 'FL_avg'].values[0] > 0.0
    ):
        FL_avg_1yr = subset2.loc[subset2['month_after_fry_emergence'] == months_later, 'FL_avg'].values[0]
        FL_SD_1yr = subset2.loc[subset2['month_after_fry_emergence'] == months_later, 'FL_SD'].values[0]
        months_later_new = months_later
    elif (
        months_later - 1 in subset2['month_after_fry_emergence'].values and
        subset2.loc[subset2['month_after_fry_emergence'] == months_later - 1, 'FL_avg'].values[0] > 0.0
    ):
        FL_avg_1yr = subset2.loc[subset2['month_after_fry_emergence'] == months_later - 1, 'FL_avg'].values[0]
        FL_SD_1yr = subset2.loc[subset2['month_after_fry_emergence'] == months_later - 1, 'FL_SD'].values[0]
        months_later_new = months_later - 1
    elif (
        months_later + 1 in subset2['month_after_fry_emergence'].values and
        subset2.loc[subset2['month_after_fry_emergence'] == months_later + 1, 'FL_avg'].values[0] > 0.0
    ):
        FL_avg_1yr = subset2.loc[subset2['month_after_fry_emergence'] == months_later + 1, 'FL_avg'].values[0]
        FL_SD_1yr = subset2.loc[subset2['month_after_fry_emergence'] == months_later + 1, 'FL_SD'].values[0]
        months_later_new = months_later + 1
    elif (
        months_later + 2 in subset2['month_after_fry_emergence'].values and
        subset2.loc[subset2['month_after_fry_emergence'] == months_later + 2, 'FL_avg'].values[0] > 0.0
    ):
        FL_avg_1yr = subset2.loc[subset2['month_after_fry_emergence'] == months_later + 2, 'FL_avg'].values[0]
        FL_SD_1yr = subset2.loc[subset2['month_after_fry_emergence'] == months_later + 2, 'FL_SD'].values[0]
        months_later_new = months_later + 2
    elif (
        months_later - 2 in subset2['month_after_fry_emergence'].values and
        subset2.loc[subset2['month_after_fry_emergence'] == months_later-+ 2, 'FL_avg'].values[0] > 0.0
    ):
        FL_avg_1yr = subset2.loc[subset2['month_after_fry_emergence'] == months_later - 2, 'FL_avg'].values[0]
        FL_SD_1yr = subset2.loc[subset2['month_after_fry_emergence'] == months_later - 2, 'FL_SD'].values[0]
        months_later_new = months_later - 2
 
    else:
        print('problem getting SD and FL 1 yr after emergence')
        plt.show()
        continue
        
    if FL_avg_1yr == 0:
        print('problem getting SD and FL 1 yr after emergence')
        plt.show()
        continue
        
 # Bootstrapping to estimate uncertainty in initial and final sizes
    num_bootstrap_samples = 1000
    initial_sizes = np.random.normal(loc=FL_avg_marentry, scale=FL_SD_marentry, size=num_bootstrap_samples)
    final_sizes = np.random.normal(loc=FL_avg_1yr, scale=FL_SD_1yr, size=num_bootstrap_samples)

    # Calculate 95% confidence intervals for both initial and final sizes
    initial_ci = np.percentile(initial_sizes, [2.5, 97.5])
    final_ci = np.percentile(final_sizes, [2.5, 97.5])

    # Plot a line between the means with confidence intervals
    x_values = [x_mo_marentry, months_later_new]
    y_values = [np.mean(initial_sizes), np.mean(final_sizes)]
    coefficients = np.polyfit(x_values, y_values, 1)
    trend_line = np.poly1d(coefficients)

    # Plot the line
    ax.plot(x_values, trend_line(x_values), linestyle='--', color='orange', label='Linear Trend')

    # Plot confidence intervals
#     ax.fill_between(x_values, initial_ci[0], initial_ci[1], alpha=0.2, color='blue', label='Initial 95% CI')
#     ax.fill_between(x_values, final_ci[0], final_ci[1], alpha=0.2, color='orange', label='Final 95% CI')

    
        
#     # Plot a line between the two sizes with confidence intervals
#     x_values = [x_mo_marentry, months_later]
#     y_values = [FL_avg_marentry, FL_avg_1yr]
#     coefficients = np.polyfit(x_values, y_values, 1)
#     trend_line = np.poly1d(coefficients)

#     # Plot the line
#     ax.plot(x_values, trend_line(x_values), linestyle='--', color='orange', label='Linear Trend')

#     # Plot confidence intervals
#     lower_bound = FL_avg_1yr - 1.96 * FL_SD_1yr
#     upper_bound = FL_avg_1yr + 1.96 * FL_SD_1yr
#     ax.fill_between(x_values, trend_line(x_values) - 1.96 * FL_SD_1yr, trend_line(x_values) + 1.96 * FL_SD_1yr, alpha=0.2, color='orange', label='95% CI')



    plt.show()
```
