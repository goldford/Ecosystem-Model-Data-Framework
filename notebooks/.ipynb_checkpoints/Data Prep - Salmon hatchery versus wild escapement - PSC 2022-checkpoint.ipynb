"This dataset provides information on enhancement level for salmon Conservation Units within the Nass, Skeena, Central Coast, Fraser, and Vancouver Island & Mainland Inlets regions of British Columbia. Enhancement Level is a measure of the extent of enhancement (e.g. from hatcheries) that has occurred within a Conservation Unit (CU), over the past two generations, and is based on methods developed by Fisheries and Oceans Canada. CU-level enhancement is calculated as the weighted average of the enhancement level of individual spawning streams within a CU with a moderate to high degree of enhancement. Enhancement level rankings are as follows: None, Low, Moderate, and High."

```python
import plotly_express as px
import pandas as pd
import altair as alt
import vega as v

#alt.renderers.enable('jupyterlab')
alt.renderers.enable('notebook')
```

```python
f_data = "Feb_3_2022_17_58_2_QueryResults.csv"
localpath_data = "C:/Users/Greig/Sync/6. SSMSP Model/Model Greig/Data/1. Salmon/All Species Hatchery vs Wild PSF 2022/MODIFIED/"
data1 = pd.read_csv(localpath_data + f_data ,parse_dates=['year'])
data1.columns
```

```python
data1['location'].unique()
```

```python
# drop WCVI data
data1 = data1.loc[data1['location']!="West Vancouver Island-Nootka and Kyuquot (Fall x-1)"]
data1 = data1.loc[data1['location']!="West Vancouver Island-North (Fall x-1)"]
data1 = data1.loc[data1['location']!="West Vancouver Island-South (Fall x-1)"].reset_index()
```

```python
chin_data = data1.loc[data1['species']=="Chinook"]
chin_data = chin_data.loc[chin_data['parameter']!="total_run"].reset_index()
#chin_data1 = chin_data.groupby(['location','RELEASE_YEAR','RELEASE_STAGE_NAME'])['TotalRelease','BIOMASS_MT'].sum().reset_index()
```

```python
source1 = chin_data

alt.Chart(source1).mark_bar().encode(
    x='year',
    y='sum(datavalue)',
    color='parameter'
)
```

```python
# group by stream-type ocean-type run
# most stream-type populations are interior fraser and 5-2
chin_data["runtype_GO"] = "ocean"
chin_data.loc[chin_data['location'] == 'Fraser Canyon-Nahatlatch (Spring 5-2)', 'runtype_GO'] = 'stream'
chin_data.loc[chin_data['location'] == 'Middle Fraser River-Portage (Fall 5-2)', 'runtype_GO'] = 'stream'
chin_data.loc[chin_data['location'] == 'Middle Fraser River (Spring 5-2)', 'runtype_GO'] = 'stream'
chin_data.loc[chin_data['location'] == 'Middle Fraser River (Summer 5-2)', 'runtype_GO'] = 'stream'
chin_data.loc[chin_data['location'] == 'North Thompson River (Spring 5-2)', 'runtype_GO'] = 'stream'
chin_data.loc[chin_data['location'] == 'North Thompson River (Summer 5-2)', 'runtype_GO'] = 'stream'
chin_data.loc[chin_data['location'] == 'South Thompson River (Summer 5-2)', 'runtype_GO'] = 'stream'
chin_data.loc[chin_data['location'] == 'Upper Fraser River (Spring 5-2)', 'runtype_GO'] = 'stream'
chin_data.loc[chin_data['location'] == 'Lower Fraser River (Spring 5-2)', 'runtype_GO'] = 'stream'
chin_data.loc[chin_data['location'] == 'Lower Fraser River (Summer 5-2)', 'runtype_GO'] = 'stream'
```

```python
chin_data_stream = chin_data.loc[chin_data['runtype_GO']=="stream"]
chin_data_ocean = chin_data.loc[chin_data['runtype_GO']=="ocean"]
```

```python
source1 = chin_data_stream

alt.Chart(source1).mark_bar().encode(
    x='year',
    y='sum(datavalue)',
    color='parameter'
)
```

```python
source1 = chin_data_ocean

alt.Chart(source1).mark_bar().encode(
    x='year',
    y='sum(datavalue)',
    color='parameter'
)
```

```python
# calculate annual percent by year and location
#chin_data_wide = df.groupby('location, year, runtype_GO').

chin_grp = chin_data.groupby(['year','runtype_GO','parameter'])['datavalue'].sum().reset_index()
#df2=df.pivot(index='countries', columns='metrics', values='values')
```

```python
chin_grp_strm = chin_grp.loc[chin_grp['runtype_GO']=="stream"]
chin_grp_strm_piv=chin_grp_strm.pivot(index='year', columns='parameter', values='datavalue')
chin_grp_strm_piv['perc_hatch'] = chin_grp_strm_piv['hatchery'] / (chin_grp_strm_piv['hatchery'] + chin_grp_strm_piv['wild'])
chin_grp_strm_piv['runtype_GO'] = "stream"
chin_grp_strm_piv = chin_grp_strm_piv.reset_index()

chin_grp_oc = chin_grp.loc[chin_grp['runtype_GO']=="ocean"]
chin_grp_oc_piv=chin_grp_oc.pivot(index='year', columns='parameter', values='datavalue')
chin_grp_oc_piv['perc_hatch'] = chin_grp_oc_piv['hatchery'] / (chin_grp_oc_piv['hatchery'] + chin_grp_oc_piv['wild'])
chin_grp_oc_piv['runtype_GO'] = "ocean"
chin_grp_oc_piv = chin_grp_oc_piv.reset_index()

# merge again
chin_grp_piv = chin_grp_strm_piv.append(chin_grp_oc_piv)
```

```python
chin_grp_piv
```

```python
# drop data prior to 1983
source1 = chin_grp_piv.loc[(chin_grp_piv['year']>'1982-01-01')&(chin_grp_piv['runtype_GO']=='ocean')]


alt.Chart(source1, title="Ocean-type Chinook Escapement Hatchery Proportion").mark_line().encode(
    x='year',
    y=alt.Y('perc_hatch', title='Prop. Hatchery')
)
```

```python
# drop data prior to 1983
source1 = chin_grp_piv.loc[(chin_grp_piv['year']>'1994-01-01')&(chin_grp_piv['runtype_GO']=='stream')]


alt.Chart(source1, title="Stream-type Chinook Escapement Hatchery Proportion").mark_line().encode(
    x='year',
    y='sum(perc_hatch)'
)
```
